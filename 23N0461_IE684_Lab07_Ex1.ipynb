**Exercise 1 :For an example (x, y) ∈ X × Y, assume z = $yw^T x$. Then note that the loss functions $L_{h}$, $L_{l}$ and $L_{sh}$ can be equivalently written as Gh(z), Gl(z), Gsh(z). Write the loss functions Gh(z), Gl(z), Gsh(z) as the functions of z.Plot these loss functions Gh(z), Gl(z) and Gsh(z) where z takes the values on the real line [−∞, ∞]. Distinguish the loss functions using different colors.Comment on the behavior of respective loss functions with respect to z.**

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the loss functions
def Gh(z):
    return np.maximum(0, 1 - z)

def Gl(z):
    return np.log(1 + np.exp(-z))

def Gsh(z):
    return (np.maximum(0, 1 - z))**2

# Generate z values
z_values = np.linspace(-10, 10, 100)

# Plot the loss functions
plt.figure(figsize=(12,8))
plt.plot(z_values, Gh(z_values), label='Hinge Loss')
plt.plot(z_values, Gl(z_values), label='Logistic Loss')
plt.plot(z_values, Gsh(z_values), label='Squared Hinge Loss')

# Add labels and legend
plt.xlabel('z')
plt.ylabel('Loss')
plt.legend()
plt.grid()


# Show the plot
plt.show()
```

**Observation :**
              
              (1) All the three functions values tends to zero after 1.
              
              (2) Squared hinge loss has maximum values compare to hinge loss and logistic loss.
              
              (3) Logistic loss and Hinge loss has somewhat same values initially.
