#Exercise 1 (15 marks) Plot the good-looking visuals of level sets for each of the following functions in (1) to (4)-:

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f1(x1, x2):
    return 100 * x2**2 + 100 * x1**4 - 200 * x1**2 * x2 + x1**2 - 2 * x1 + 1

# Create a meshgrid for x1 and x2
x1 = np.linspace(-4, 4, 400)
x2 = np.linspace(-4, 4, 400)
X1, X2 = np.meshgrid(x1, x2)

# Evaluate the function on the meshgrid
Z1 = f1(X1, X2)

# Plot level sets
plt.contour(X1, X2, Z1, levels=np.logspace(0, 5, 50), cmap='viridis')
plt.title('Level Sets of Function 1')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

```python
# Define the function
def f2(x1, x2):
    return x1**4 + x2**4 - 20 * x1**3  - 20 * x2**3 + 100 * x1**2 + 100 * x2**2

# Evaluate the function on the meshgrid
Z2 = f2(X1, X2)

# Plot level sets
plt.contour(X1, X2, Z2, levels=np.logspace(-2, 3, 50), cmap='plasma')
plt.title('Level Sets of Function 2')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

```python
# Define the function
def f3(x1, x2):
    return -np.cos(x1) * np.cos(x2) * np.exp(-((x1 - np.pi)**2 + (x2 - np.pi)**2) )

# Evaluate the function on the meshgrid
Z3 = f3(X1, X2)

# Plot level sets
plt.contour(X1, X2, Z3, levels=np.linspace(-2, 2, 50), cmap='cividis')
plt.title('Level Sets of Function 3')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

```python
# Define the function
def f4(x1, x2):
    return 2 * x1**2 - 1.05 * x1**4 +  + x1**6/6 + x1 * x2 + x2**2

# Evaluate the function on the meshgrid
Z4 = f4(X1, X2)

# Plot level sets
plt.contour(X1, X2, Z4, levels=np.logspace(-1, 3, 50), cmap='magma')
plt.title('Level Sets of Function 4')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```
